```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as p3
from sklearn.datasets import make_swiss_roll

# #############################################################################
# Generate data (swiss roll dataset)
n_samples = 1500
noise = 0.05
X, t = make_swiss_roll(n_samples, noise)
# Make it thinner
X[:, 1] *= .5
```

```python
fig = plt.figure()
ax = p3.Axes3D(fig)
ax.view_init(7, -80)
ax.scatter(X[:,0], X[:,1], X[:,2], s=20, edgecolor='k', c=t, cmap='viridis')

#plt.title('Without connectivity constraints (time %.2fs)' % elapsed_time)
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

plt.figure(figsize=(10,5))
pca = PCA(n_components=2).fit(X.T)
plt.subplot(1,2,1)
plt.scatter(pca.components_[0], pca.components_[1], c=t, edgecolor='k')
plt.title('PCA')

tsne = TSNE(n_components=2, perplexity=60, learning_rate=1500/12).fit(X)

plt.subplot(1,2,2)
plt.scatter(tsne.embedding_[:,0], tsne.embedding_[:,1], c=t, edgecolor='k')
plt.title('tSNE')
```
